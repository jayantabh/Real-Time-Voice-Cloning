```
import os
import random
dirName = '/content/Real-Time-Voice-Cloning/LibriSpeech/train-clean-100'
listOfFiles = list()
for (dirpath, dirnames, filenames) in os.walk(dirName):
    listOfFiles += [os.path.join(dirpath, file)  for file in filenames if file.endswith(".flac")]
random.shuffle(listOfFiles)
train_list = listOfFiles[:-2000]
val_list = listOfFiles[-2000:]
```

```
with open('/content/Real-Time-Voice-Cloning/waveglow/train_files.txt', 'w') as f:
  for fname in train_list:
    f.write(fname + '\n')

with open('/content/Real-Time-Voice-Cloning/waveglow/val_files.txt', 'w') as f:
  for fname in val_list:
    f.write(fname + '\n')
```
